```python
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import tensorflow as tf

def preprocess_paired_image(image_path, img_size=(256, 256)):
    """
    Splits paired images into input (left) and target (right).
    """
    img = load_img(image_path, target_size=(img_size[0], img_size[1] * 2))  # Load full image
    img = img_to_array(img) / 255.0  # Normalize to [0, 1]
    input_img = img[:, :img_size[1], :]  # Left half
    target_img = img[:, img_size[1]:, :]  # Right half
    return input_img, target_img

# Example: Preprocess a single image from the train directory
sample_image_path = os.path.join(train_path, '387.jpg')
input_img, target_img = preprocess_paired_image(sample_image_path)

# Visualize input and target images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(input_img)
axes[0].set_title("Input (Label)")
axes[0].axis("off")
axes[1].imshow(target_img)
axes[1].set_title("Target (Real Image)")
axes[1].axis("off")
plt.show()
```

```python
import numpy as np

def load_and_preprocess_data(train_dir, img_size=(256, 256)):
    """
    Loads and preprocesses all images in the training directory.
    Returns the input (X) and target (Y) image arrays.
    """
    input_images = []
    target_images = []

    # Loop through all files in the train directory
    for file_name in os.listdir(train_dir):
        if file_name.endswith('.jpg'):
            img_path = os.path.join(train_dir, file_name)
            input_img, target_img = preprocess_paired_image(img_path, img_size)
            input_images.append(input_img)
            target_images.append(target_img)

    # Convert lists to numpy arrays
    X = np.array(input_images)
    Y = np.array(target_images)

    return X, Y

# Load and preprocess the entire dataset
X_train, Y_train = load_and_preprocess_data(train_path)

# Print the shape of the data
print(f"Shape of input images: {X_train.shape}")
print(f"Shape of target images: {Y_train.shape}")
```

```python
from tensorflow.keras import layers, models

def build_generator():
    inputs = layers.Input(shape=(256, 256, 3))

    # Encoder
    down1 = layers.Conv2D(64, (4, 4), strides=2, padding='same', activation='relu')(inputs)
    down2 = layers.Conv2D(128, (4, 4), strides=2, padding='same', activation='relu')(down1)
    down3 = layers.Conv2D(256, (4, 4), strides=2, padding='same', activation='relu')(down2)
    down4 = layers.Conv2D(512, (4, 4), strides=2, padding='same', activation='relu')(down3)

    # Bottleneck
    bottleneck = layers.Conv2D(1024, (4, 4), strides=2, padding='same', activation='relu')(down4)

    # Decoder (Upsampling)
    up1 = layers.Conv2DTranspose(512, (4, 4), strides=2, padding='same', activation='relu')(bottleneck)
    up2 = layers.Conv2DTranspose(256, (4, 4), strides=2, padding='same', activation='relu')(up1)
    up3 = layers.Conv2DTranspose(128, (4, 4), strides=2, padding='same', activation='relu')(up2)
    up4 = layers.Conv2DTranspose(64, (4, 4), strides=2, padding='same', activation='relu')(up3)

    outputs = layers.Conv2D(3, (4, 4), strides=1, padding='same', activation='tanh')(up4)

    model = models.Model(inputs, outputs)
    return model

generator = build_generator()
generator.summary()
```

```python
def build_discriminator():
    inputs = layers.Input(shape=(256, 256, 3))

    x = layers.Conv2D(64, (4, 4), strides=2, padding='same', activation='relu')(inputs)
    x = layers.Conv2D(128, (4, 4), strides=2, padding='same', activation='relu')(x)
    x = layers.Conv2D(256, (4, 4), strides=2, padding='same', activation='relu')(x)
    x = layers.Conv2D(512, (4, 4), strides=2, padding='same', activation='relu')(x)

    x = layers.Flatten()(x)
    x = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs, x)
    return model

discriminator = build_discriminator()
discriminator.summary()
```

```python
from tensorflow.keras import layers, models

def build_generator():
    inputs = layers.Input(shape=(256, 256, 3))

    # Encoder
    down1 = layers.Conv2D(64, (4, 4), strides=2, padding='same', activation='relu')(inputs)
    down2 = layers.Conv2D(128, (4, 4), strides=2, padding='same', activation='relu')(down1)
    down3 = layers.Conv2D(256, (4, 4), strides=2, padding='same', activation='relu')(down2)
    down4 = layers.Conv2D(512, (4, 4), strides=2, padding='same', activation='relu')(down3)

    # Bottleneck
    bottleneck = layers.Conv2D(1024, (4, 4), strides=2, padding='same', activation='relu')(down4)

    # Decoder (Upsampling)
    up1 = layers.Conv2DTranspose(512, (4, 4), strides=2, padding='same', activation='relu')(bottleneck)
    up2 = layers.Conv2DTranspose(256, (4, 4), strides=2, padding='same', activation='relu')(up1)
    up3 = layers.Conv2DTranspose(128, (4, 4), strides=2, padding='same', activation='relu')(up2)
    up4 = layers.Conv2DTranspose(64, (4, 4), strides=2, padding='same', activation='relu')(up3)

    # Add another Conv2DTranspose layer to increase the output size
    up5 = layers.Conv2DTranspose(3, (4, 4), strides=2, padding='same', activation='tanh')(up4) # Add this line


    #outputs = layers.Conv2D(3, (4, 4), strides=1, padding='same', activation='tanh')(up4)
    # Change the output to be up5
    outputs = up5 # Change to up5

    model = models.Model(inputs, outputs)
    return model

generator = build_generator()
generator.summary()
```

```python
# Import the necessary optimizer class
from tensorflow.keras.optimizers import Adam

# Create an instance of the optimizer (e.g., Adam)
optimizer = Adam(learning_rate=0.0002, beta_1=0.5)

# Now you can compile the discriminator
discriminator.compile(loss='binary_crossentropy',
                      optimizer=optimizer,  # Use the defined optimizer
                      metrics=['accuracy'])
```

```python
# Assuming 'generator' and 'discriminator' are defined as in previous cells

# Define the Pix2Pix model
# Input is the generator input (source image)
input_src_image = layers.Input(shape=(256, 256, 3))

# Generate fake target image
gen_out = generator(input_src_image)

# Connect the discriminator to the generated image
# (but disable training for the discriminator during this step)
discriminator.trainable = False
dis_out = discriminator(gen_out)

# Define the combined Pix2Pix model
pix2pix = models.Model(input_src_image, [gen_out, dis_out], name='pix2pix')

# Now you can compile the Pix2Pix model
pix2pix.compile(loss=['mae', 'binary_crossentropy'],
                loss_weights=[100, 1],  # weight on the L1 loss vs adversarial loss
                optimizer=optimizer)
```

```python
import numpy as np
import tensorflow as tf

def train_pix2pix(generator, discriminator, pix2pix, X_train, Y_train, epochs, batch_size):
    """
    Trains the Pix2Pix model.
    """

    # Calculate the number of batches per epoch
    num_batches = int(len(X_train) / batch_size)

    # Loop through epochs
    for epoch in range(epochs):
        # Shuffle the training data
        indices = np.arange(len(X_train))
        np.random.shuffle(indices)
        X_train = X_train[indices]
        Y_train = Y_train[indices]

        # Loop through batches
        for batch_index in range(num_batches):
            # Get the current batch of data
            start = batch_index * batch_size
            end = (batch_index + 1) * batch_size
            X_batch = X_train[start:end]
            Y_batch = Y_train[start:end]

            # Create fake images (generated images)
            generated_images = generator.predict(X_batch)

            # Train discriminator (real vs. fake)
            discriminator_loss_real = discriminator.train_on_batch(Y_batch, np.ones((batch_size, 1)))  # Real images labeled as 1
            discriminator_loss_fake = discriminator.train_on_batch(generated_images, np.zeros((batch_size, 1)))  # Fake images labeled as 0

            # Calculate average discriminator loss
            discriminator_loss = 0.5 * np.add(discriminator_loss_real, discriminator_loss_fake)

            # Train the Pix2Pix model (generator part)
            # Real images labeled as 1 to fool the discriminator
            pix2pix_loss = pix2pix.train_on_batch(X_batch, [Y_batch, np.ones((batch_size, 1))])

            # Display training progress (optional)
            if batch_index % 100 == 0:
                print(f"Epoch: {epoch + 1}, Batch: {batch_index}/{num_batches}, "
                      f"D loss: {discriminator_loss[0]}, Acc: {discriminator_loss[1]}, "
                      f"G loss: {pix2pix_loss[0]}")
```
